# Plano de Análise
 - Vamos buscar relações que podem estar associadas a taxa de desemprego
 
 - Identificar onde existe a maior e a menor taxa de desemprego
 
 - Analisaremos a variação de queda e aumento de desemprego nos anos recentes
 

```python
# Primeiro iremos visualizar todos os países do DataFrame.
df_Paises = df['Country Name']
pd.set_option('display.max_row',236)

df_Paises
```

```python
# Deletando Colunas que não são países e nem relevantes para a análise, pelo Index.
df_Paises.drop([82,193,148,89,90,91,92,93,94,229,220,213,208,210,212,202,203,190,
                     167,140,124,121,122,118,117,111,56,54,174,158,85,66,119,160])
```

```python
# Agrupar nossos Dados primeiro por regiões na América: América Central, América do Sul, América Do Norte
df = df.set_index('Country Name')
America_Central = df.loc[['Belize','Costa Rica','Cuba','El Salvador','Guatemala','Honduras','Nicaragua',
                      'St. Lucia','Bahamas, The','Barbados','Panama','Dominican Republic',
                      'Jamaica','St. Vincent and the Grenadines','Trinidad and Tobago','Haiti']]

America_Sul = df.loc[['Brazil','Bolivia','Argentina','Chile','Colombia','Ecuador','Guyana','Peru',
                      'Paraguay','Suriname','Uruguay','Venezuela, RB']]

America_Norte = df.loc[['United States','Mexico','Canada']]
```

```python
# Vamos realizar um levantamento que irá nos mostrar a taxa de desemprego geral no ano mais recente-2021,
# Lembrando que infelizmente não possuímos todos os países no nosso caso em questão.
America_Central['Country Name'] = America_Central.index

plt.figure(figsize = (12,9))
America_Central.plot( x = 'Country Name',
                      y = '2021',
                      kind='bar',
                      stacked = True)
plt.show()
```

```python
America_Sul['Country Name'] = America_Sul.index

plt.figure(figsize = (12,9))
America_Sul.plot( x = 'Country Name',
                      y = '2021',
                      kind='bar',
                      stacked = True)
plt.show()
```

```python
America_Norte['Country Name'] = America_Norte.index

plt.figure(figsize = (8,6))
America_Norte.plot( x = 'Country Name',
                      y = '2021',
                      kind='bar',
                      stacked = True)
plt.show()
```

- Conseguimos visualizar que entre as 3 Regiões da América, e entre todos os países,
 a menor taxa de desemprego se encontra na América Central, no país de Cuba com cerca de 3% de taxa de desemprego.
 E a maior taxa também se encontra na América Central, no país de São Vicente e Granadinas com aproximadamente 22% de  taxa de desemprego, no ano de 2021.

```python
#Vamos criar uma nova coluna que nos mostrará a diferença do da taxa de desemprego entre 10 anos (2011-2021)
America_Central['Variação_Década'] = America_Central['2011'] - America_Central['2021']
America_Central[['Variação_Década']]
```

```python
#Vamos criar uma nova coluna que nos mostrará a diferença do da taxa de desemprego entre 10 anos (2011-2021)
America_Norte['Variação_Década'] = America_Norte['2011'] - America_Norte['2021']
America_Norte[['Variação_Década']]
```

```python
#Vamos criar uma nova coluna que nos mostrará a diferença do da taxa de desemprego entre 10 anos (2011-2021)
America_Sul['Variação_Década'] = America_Sul['2011'] - America_Sul['2021']
America_Sul[['Variação_Década']]
```

- Podemos visualizar que no decorrer da década,Aproximadamente 75% dos países não sofreram um redução na taxa de desemprego
  uma ótima notícia. O país com maior taxa em 10 anos foi o Estados Unidos,Enquanto que a maior queda da taxa de desemprego se encontra na Costa Rica,Já países como Canadá,Belize,Cuba mantiveram a taxa de desemprego na mesma média em 10 anos.
